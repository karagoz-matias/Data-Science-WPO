Now, plot separate boxplots corresponding to the first 14 days of the columns "TMAX", "TMIN" and "PRCP" and give the plot a title "BOX plot with outliers". Boxplot is a method of matplotlib.pyplot just like subplots is. Check the documentation here: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html.

```python
#1 and 2
import matplotlib.pyplot as plt
import pandas as pd
weather_data = pd.read_csv("seattleWeather_1948-2017.csv")
x = weather_data["DATE"][0:14]
y0 = weather_data["PRCP"][0:14]
y1 = weather_data["TMAX"][0:14]
y2 = weather_data["TMIN"][0:14]
fig, ax= plt.subplots(3, figsize=(15,10), sharex = True)
ax[0].set_title('Weather in Seattle')
ax[0].scatter(x,y0, color="blue", marker = 'x')
ax[0].set_ylabel("PRCP")
ax[1].scatter(x,y1, color="red")
ax[1].set_ylabel("TMAX")
ax[2].scatter(x,y2, color="green")
ax[2].set_ylabel("TMIN")
plt.show()


#3
fig, ax= plt.subplots(3,figsize=(15,10), sharex = True )
ax[0].boxplot(y0)
ax[0].set_title('PRCP')
ax[1].boxplot(y1)
ax[1].set_title('TMAX')
ax[2].boxplot(y2)
ax[2].set_title('TMIN')
plt.show()
```
