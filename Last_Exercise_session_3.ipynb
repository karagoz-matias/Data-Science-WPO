**Option 2:** **Replace** the missing values by the 'true values' or by imputing values based on the non-missing data


Remember from HOC 2 that simply removing data can induce negative side effects on the performance of your machine learning algorithm. In order to avoid these, one can rely on the "SimpleImputer" function provided by Scikit-learn: https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html. This function will allow you to replace the missing values with the mean, median, most-frequent values from that column, or a constant value.

In order to use the SimpleImputer function, we need to create a SimpleImputer-instance first (actually, the use of this function has a lot in common with the use of thedecisiontree functions used before. )

```python
from sklearn.impute import SimpleImputer

NA_imputer_1=SimpleImputer(strategy = 'median')  #Use this object if you want to replace each attibute's missing values with the median of that attribute

NA_imputer_2=SimpleImputer(strategy = 'mean')  #Use this object if you want to replace each attibute's missing values with the mean value of that attribute

NA_imputer_3=SimpleImputer(strategy = 'most_frequent') #Use this object if you want to replace each attibute's missing values with the most frequent value of that attribute :  for categorical features

NA_imputer_4=SimpleImputer(strategy = 'constant', fill_value = 3) #Use this object if you want to replace each attibute's missing values with value '3' or any other value assigned to the parameter 'fill_value'.
```

Once the instances are created we can use the “fit” method to the appropriate column to let the “SimpleImpute” compute the statistical properties of that column, needed for imputation.

```python
from sklearn.impute import SimpleImputer

NA_imputer_1=SimpleImputer(strategy = 'median')  #Use this object if you want to replace each attibute's missing values with the median of that attribute

NA_imputer_2=SimpleImputer(strategy = 'mean')  #Use this object if you want to replace each attibute's missing values with the mean value of that attribute

NA_imputer_3=SimpleImputer(strategy = 'most_frequent') #Use this object if you want to replace each attibute's missing values with the most frequent value of that attribute :  for categorical features

NA_imputer_4=SimpleImputer(strategy = 'constant', fill_value = 3)

NA_imputer_1.fit(housing.drop('ocean_proximity', axis = 1))                # we apply the fit method on the whole dataset (the .fit method is telling to the computer to look at the data without changing it yet), except from the ocean proximity column as it contains non-numerical values we still have to convert (which we do later in the notebook)

NA_imputer_2.fit(housing.drop('ocean_proximity', axis = 1))

NA_imputer_3.fit(housing.drop('ocean_proximity', axis = 1))

NA_imputer_4.fit(housing.drop('ocean_proximity', axis = 1))
```

## 2. Converting categorical attributes to numerical/one-hot-encoded values

Most Machine Learning algorithms prefer to work with numbers. However, some datasets contain categorical attributes.

 In order to convert text to numbers, we can use Scikit-Learn's "OrdinalEncoder" and "OneHotEncoder" classes. As an illustration, we will apply its methods to the housing dataset.  In general, there are two strategies to eliminate text and convert it to number:

* Ordinal Encoder: each unique category value is assigned an integer value. Imposes ordinal relationship between categories (!) (category 1 is closer to 2 than 3, categories are ranked)
* One-Hot-Encoder: the integer encoded variable is removed and one new binary variable is added for each unique integer value in the variable (no ordinal relationship induced, no natural ranking).

Let's print the "ocean_proximity" column first.

```python
housing
```

**Ordinal Encoder**

First, we import the OrdinalEncoder function and create an object with default parameters. For more information about this function, check: https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OrdinalEncoder.html.

```python
from sklearn.preprocessing import OrdinalEncoder
```

Then, we need to create an instance of the OrdinalEncoder object:

```python
OE = OrdinalEncoder()
```

Fit and transform the dataset. Again, if you fit and transform on the same data, you can use “fit_transform()” in one step. Convert the result to a pandas dataframe.

```python
ordinal_columns = housing[['ocean_proximity']]
Ordinal_encoded = pd.DataFrame(OE.fit_transform(ordinal_columns), columns=['ocean_proximity'])
```

Join the newly encoded columns with the original dataframe. Do not forget to drop columns to avoid doubles.

```python
full_encoded = pd.concat([housing.drop(['ocean_proximity'], axis = 1), Ordinal_encoded], axis =1)
full_encoded
```

**One-Hot-Encoder**

One issue with the representation presented above is that ML algorithms will assume that two nearby values are more similar than two distant values. This may be fine in some cases (e.g., for ordered categories such as “bad”, “average”, “good”, “excellent”), but it is obviously not always the case. To fix this issue, a common solution is to create one binary attribute per category: one attribute equal to 1 when the category is “<1H OCEAN” (and 0 otherwise), another attribute equal to 1 when the category is “INLAND” (and 0 otherwise), and so on. This is called one-hot encoding, because only one attribute will be equal to 1 (hot), while the others will be 0 (cold). The new attributes are sometimes called dummy attributes. Scikit-Learn provides a OneHotEn coder class to convert categorical values into one-hot vectors. More information about this class: https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html

First, import the OneHotEncoder from the sklearn.preprocessing package.

```python
from sklearn.preprocessing import OneHotEncoder
```

Then, create an instance of the OneHotEncoder.

```python
OHE = OneHotEncoder()
```

Use the fit method of the OneHotEncoder and apply it to the "ocean_proximity" column. Before doing so, create a variable that contains only this column.

```python
OHE_columns = housing[['ocean_proximity']]
OHE.fit(OHE_columns)
```

Now, use the “.categories_” method to see which category names the OHE detected

```python
OHE.categories_
```

If all categories are properly recognized, use the "transform" method to actually apply the OHE.

Finally, convert the result to an array by using the "toarray()" method. Then convert that array to a pandas dataframe, specify the column names by copying the names from the array (point 3).

```python
OHE_encoded = pd.DataFrame(OHE.fit_transform(OHE_columns).toarray(), columns=['<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN'])
```

To end with, we still have to merge these new columns with the rest of the dataset. To do so, we use pandas "concat" function.

```python
full_enc = pd.concat([housing.drop(['ocean_proximity'], axis = 1), OHE_encoded], axis =1)
full_enc
```
